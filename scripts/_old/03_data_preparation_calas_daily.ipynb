# 03 - Calas Data Preparation - Daily Aggregation

Aggregates individual fishing event records (calas) to daily totals per vessel, company, and season, keeping mean coordinates.

**Steps:**
1. Load the full calas dataset produced by notebook 02
2. Group by date, season, company, and vessel
3. Sum catch tonnage and average lat/lon per group
4. Save the daily catch table with coordinates to outputs

```python
import pandas as pd
from pathlib import Path
```

```python
calas_path = Path("/home/jupyter-daniela/suyana/peru_production/outputs/")

df_calas = pd.read_csv(calas_path / "calas_all_data.csv", low_memory=False)
df_calas['fecha_cala'] = pd.to_datetime(df_calas['fecha_cala'], errors='coerce')
df_calas = df_calas.rename(columns={'latitud ': 'latitud'})
```

```python
df_calas_daily = (
    df_calas.groupby(['fecha_cala', 'temporada', 'empresa', 'nave'])
    .agg({
        'declarado_tm': 'sum',
        'latitud': 'mean',
        'longitud': 'mean'
    })
    .reset_index()
    .rename(columns={
        'fecha_cala':  'date',
        'temporada':   'season',
        'empresa':     'company',
        'nave':        'vessel',
        'declarado_tm':'catch_tm',
        'latitud':     'lat',
        'longitud':    'lon',
    })
)
```

```python
df_calas_daily.to_csv(calas_path / "calas_daily_with_coordinates.csv", index=False)
```
